```python
from google_play_scraper import Sort, reviews_all
from textblob import TextBlob
import pandas as pd

# Mengambil semua ulasan dari aplikasi WhatsApp
scrapreview = reviews_all(
    'com.whatsapp',        # ID aplikasi
    lang='en',             # Bahasa ulasan
    country='us',          # Negara
    sort=Sort.MOST_RELEVANT,  # Urutan ulasan
    count=500000           # Jumlah maksimum ulasan
)

# Membuat DataFrame dari hasil ulasan
reviews_df = pd.DataFrame(scrapreview)

def analyze_sentiment(review):
    analysis = TextBlob(review)
    polarity = analysis.sentiment.polarity
    if polarity > 0:
        return 'positive'
    elif polarity == 0:
        return 'neutral'
    else:
        return 'negative'

# Menambahkan kolom sentimen ke DataFrame
reviews_df['sentiment'] = reviews_df['content'].apply(analyze_sentiment)

# Menampilkan 5 baris pertama dari DataFrame
print(reviews_df.head())

# Menghitung jumlah tiap kelas sentimen
sentiment_counts = reviews_df['sentiment'].value_counts()
print("Sentiment Counts:")
print(sentiment_counts)

# Menyimpan DataFrame ke file CSV dengan pengaturan escapechar dan quotechar
reviews_df.to_csv('whatsapp_reviews.csv', index=False, escapechar='\\', quotechar='"')
```

```python
# Memuat kembali DataFrame dari file CSV
reviews_df = pd.read_csv('whatsapp_reviews.csv', escapechar='\\', quotechar='"')

# Filter ulasan yang memiliki lebih dari atau sama dengan 5 kata
filtered_reviews_df = reviews_df[reviews_df['content'].apply(lambda x: len(x.split()) >= 5)]

# Menyimpan DataFrame yang telah difilter ke file CSV baru
filtered_reviews_df.to_csv('whatsapp_mentah.csv', index=False, escapechar='\\', quotechar='"')
```
